Transformation

```python
query = """
SELECT
    ROW_NUMBER() OVER (ORDER BY ci.customer_id) AS customer_key,
    ci.customer_id,
    ci.customer_number,
    ci.first_name,
    ci.last_name,
    la.country,
    ci.marital_status,
    CASE
        WHEN ci.gender <> 'n/a' THEN ci.gender
        ELSE COALESCE(ca.gender, 'n/a')
    END AS gender,
    ca.birth_date AS birthdate,
    ci.created_date AS create_date
FROM lakehouse_project.silver.crm_customers ci
LEFT JOIN lakehouse_project.silver.erp_customers ca
    ON ci.customer_number = ca.customer_number
LEFT JOIN lakehouse_project.silver.erp_customer_location la
    ON ci.customer_number = la.customer_number
"""
df = spark.sql(query)
```

```python
display(df.limit(10))
```

Writing Gold table

```python
df.write.mode("overwrite").format("delta").saveAsTable("lakehouse_project.gold.dim_customers")
```
