Transformation

```python
query = """
SELECT
    sd.order_number,
    pr.product_key,
    cu.customer_key,
    sd.order_date,
    sd.ship_date,
    sd.due_date,
    sd.sales_amount,
    sd.quantity,
    sd.price
FROM lakehouse_project.silver.crm_sales sd
LEFT JOIN lakehouse_project.gold.dim_products pr
    ON sd.product_number = pr.product_number
LEFT JOIN lakehouse_project.gold.dim_customers cu
    ON sd.customer_id = cu.customer_id;
"""
df = spark.sql(query)
```

```python
display(df.limit(10))
```

Writing gold table

```python
df.write.mode("overwrite").format("delta").saveAsTable("lakehouse_project.gold.fact_sales")
```
